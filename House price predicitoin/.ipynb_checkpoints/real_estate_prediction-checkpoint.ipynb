## Model creation process:
    1. import data
    2. split train test then their feautres and labels
    3. Pipeline the data:
        1. impute, fill missing data using SimpleImputer
        2. scale the data using StandardScaler
    4. Create model piplined train feature and train label
    5. Evalute and after low error, proceed otherwise use differnet model
    6. Test model using pipelined test feature
    7. Save model

### filling the null values

```python
housing.RM.describe()
```

```python
housing.RM.fillna(housing.RM.mean(),inplace=True)
```

```python
housing.skew()
```

```python
housing.isnull().sum()
```

```python
X = housing.drop("MEDV", axis=1)
y = housing.MEDV.copy()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=324) 
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

# Scaling the data
    Scaling : transform the data such that mean will be 0 and std will be 1
    

```python
#standard scaler will replace all the value such that standard deviation of each column will be one
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

### checking the std and mean of scaled data

```python
X_df = pd.DataFrame(X_train,columns=housing.columns[:-1])
X_df
```

```python
X_df.std()
```

```python
X_df.mean()
```

```python
X_df.skew()
```

## Creating model
    Model tried:
        1. LinearRegressor
        2. DecisionTreeRegressor
        3. RandomForestRegressor(best)

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
model = [LinearRegression(),DecisionTreeRegressor(),RandomForestRegressor()]
for i in model:
    i.fit(X_train, y_train)
```

## Evaluting all model using rmse on train data

```python
from sklearn.metrics import mean_squared_error
housing_predictions = []
for c,i in enumerate(model):
    housing_predictions.append(i.predict(X_train)) 
    mse = mean_squared_error(y_train, housing_predictions[c])
    rmse = np.sqrt(mse)
    print(f"{i} RMSE value : {rmse}")
```

### NOTE
    Decision tree is over fitted as the error is 0
    Random forest model is better as the error is lower than linear regression model

```python
sns.regplot(X_df.LSTAT,housing_predictions[0], color="red")
sns.regplot(X_df.LSTAT,y_train, color="blue")
```

```python
sns.regplot(X_df.LSTAT,housing_predictions[1], color="red")
sns.regplot(X_df.LSTAT,y_train, color="blue")
## note how all the point is overlapped making it overfitted
```

```python
sns.regplot(X_df.LSTAT,housing_predictions[2], color="red")
sns.regplot(X_df.LSTAT,y_train, color="blue")
```

## Testing model on test data 
    

```python
X_test_prepared = scaler.transform(X_test)
final_predictions = []
for i in model:
    final_predictions.append(i.predict(X_test_prepared)) 
```

```python
##evalulatind test data
for i,j in zip(final_predictions,model):
    final_mse = mean_squared_error(y_test, i)
    final_rmse = np.sqrt(final_mse)
    print(f"{j}----->{final_rmse}")
```

#### NOTE
     Random forest has lowest error on test data
     Since Decision tree was overfitted it has more error
